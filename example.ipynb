Fill in the email and API token with sufficient permissions (at least reading account, workspaces, and repository).

```python
# Provide your information here
email = ""
token = ""
```

First, we get all workspaces and print the results.

If you run this file not within the bucketcloner source directory, you might need to adjust the import statement.

```python
import bucketcloner.main as bucketcloner

workspaces = bucketcloner.list_bitbucket_workspaces(email, token)
for w in workspaces:
    print(f"{w['name']} ({w['slug']}) - {w['url']}")
```

Use the last workspace from the list and retrieve its repositories. The repositories will be downloaded in a folder with the name of the _slug_ of the workspace.

```python
from pathlib import Path

from bucketcloner.main import AuthMode, BucketClonerConfig

workspace = workspaces[-1]
config = BucketClonerConfig(
    email=email,
    token=token,
    skip_existing=True,
    refresh=False,
    clone_into_project_folders=False,
    base_folder=Path().cwd(),
    auth_mode=AuthMode.HTTPS,
    ssh_key=None,
)


bucketcloner.clone_bitbucket(email, token, config)
```

Now we check if the workspace was checked out and print all repositories

```python
repos = list(Path(workspace["slug"]).iterdir())
print(repos)
```

Finally, we delete the local workspace folder.

```python
import contextlib
import shutil

with contextlib.suppress(FileNotFoundError):
    shutil.rmtree(workspace["slug"])
```
